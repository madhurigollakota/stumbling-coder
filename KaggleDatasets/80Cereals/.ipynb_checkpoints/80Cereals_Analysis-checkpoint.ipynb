```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
Cereal=pd.read_csv('cereal.csv')
```

```python
Cereal.head(5)
```

```python
#Check for null values count
Cereal.info(verbose=True)
```

```python
#Convert all the columns to the same measurement unit. In the dataset given, sodium and potassium are 
#in milligrams while the other columns are in grams.
Cereal.sodium=Cereal.sodium/1000
Cereal.potass=Cereal.potass/1000
```

```python
Cereal.describe()
```

```python
#Replace negative values with 0 - Carbo, sugars,potasium
print(Cereal.loc[Cereal.carbo<0])
print(Cereal.loc[Cereal.potass<0])
print(Cereal.loc[Cereal.sugars<0])

Cereal.loc[Cereal.carbo<0,'carbo']=0
Cereal.loc[Cereal.sugars<0,'sugars']=0
Cereal.loc[Cereal.potass<0,'potass']=0

#Another way of doing the above operation
#Cereal.carbo=[ 0 if i<0 else i for i in Cereal.carbo]
#Cereal.sugars=[ 0 if i<0 else i for i in Cereal.sugars]
#Cereal.potass=[ 0 if i<0 else i for i in Cereal.potass]
```

```python
Cereal.describe()
```

```python
#Data analysis by Manufacture
mfr=Cereal.groupby('mfr').mean().reset_index()

#Aggregation by calories
plt.figure(1,figsize=(16,7))
plt.subplot(2,3,1)
plt.title('Aggregation by calories')
plt.pie(mfr.calories,labels=mfr.mfr,autopct='%1.1f%%')

#Aggregation by sugars
#plt.figure(2,figsize=(16,7))
plt.subplot(2,3,2)
plt.title('Aggregation by sugars')
plt.pie(mfr.sugars,labels=mfr.mfr,autopct='%1.1f%%')

plt.subplot(2,3,3)
plt.title('Aggregation by fiber')
plt.pie(mfr.fiber,labels=mfr.mfr,autopct='%1.1f%%')

plt.subplot(2,3,4)
plt.title('Aggregation by fat')
plt.pie(mfr.fat,labels=mfr.mfr,autopct='%1.1f%%')

plt.subplot(2,3,5)
plt.title('Aggregation by protein')
plt.pie(mfr.protein,labels=mfr.mfr,autopct='%1.1f%%')

plt.subplot(2,3,6)
plt.title('Aggregation by carbs')
plt.pie(mfr.carbo,labels=mfr.mfr,autopct='%1.1f%%')
```
